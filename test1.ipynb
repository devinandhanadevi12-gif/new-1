```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

sns.set_theme(style="darkgrid")

```

```python
df = pd.read_csv('hours_pass.csv')
print("Dataset:")
print(df.head())

plt.figure(figsize=(8, 5))
plt.scatter(df['hours'], df['pass'], c=df['pass'], cmap='bwr', edgecolor='k')
plt.title('Hours Studied vs Pass')
plt.xlabel('Hours Studied')
plt.ylabel('Pass(1) / Fail(0)')
plt.title('Hours Studied vs Pass')
plt.show()
```

```python
x = df[['hours']] # input
y = df['pass'] # output

model = LogisticRegression()
model.fit(x, y)
print("Model trained successfully.")
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['hours'], df['pass'], c=df['pass'], cmap='bwr', edgecolor='k')

x_test_range = np.linspace(0, 10, 100).reshape(-1, 1)
y_prob = model.predict_proba(x_test_range)[:, 1]
plt.plot(x_test_range, y_prob, color='green', linewidth=3, label='logistic curve')
plt.axhline(0.5, color='red', linestyle='--', label='Decision Boundary (0.5)')  
plt.xlabel('Hours Studied')
plt.ylabel('Pass Probability')  
plt.title('Logistic Regression sigmoid curve')
plt.legend()    
plt.show()
```

```python
# get user input
try:
    user_hours = float(input("Enter hours studied: "))
    x_user = np.array([[user_hours]])
    prediction = model.predict(x_user)[0]
    probability = model.predict_proba(x_user)[0][1]

    if prediction == 1:
     print(f"\nResult: PASS")
     print(f"\nResult: FAIL\nConfidence: {probability*100:.2f}%")
    else:
     print(f"\nResult: FAIL")
     print(f"\nResult: PASS\nConfidence: {(1-probability)*100:.2f}%")   
except ValueError:
    print("Invalid input. Please enter a numeric value for hours studied.") 

```
